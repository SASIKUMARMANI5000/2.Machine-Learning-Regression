```python
# pandas is a library and it contains the functions related to the database
```

```python
import pandas as pd
```

```python
# pandas(Pd) using called to dataset of particular database (.csv_file)
```

```python
dataset=pd.read_csv("Salary_Data.csv")
```

```python
#To run the dataset 
```

```python
dataset
```

```python
# To Called input(independent) dataset only
```

```python
independent=dataset[["YearsExperience"]]
```

```python
# To run the independent
```

```python
independent
```

```python
# To called output(dependent) dataset only
```

```python
dependent=dataset[["Salary"]]
```

```python
# To run the dependent
```

```python
dependent
```

```python
# Spliting the train_set and test_set to using sklearn(folder)-function(.)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(independent, dependent, test_size=0.30, random_state=0)
```

```python
# The model creating process ofSubstituting values in regressor(.fit) to calculate (weight and bais) using 'LinearRegression Algorithms'
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(X_train,Y_train)
```

```python
# To run for check the value of 'weight/slope' using variable (.coef_)
```

```python
weight=regressor.coef_
weight
```

```python
# To run for check the value of 'bais' using variable (.intercept_)
```

```python
bais=regressor.intercept_
bais
```

```python
# Evaluating the model to cross chcek using r2_score
```

```python
y_pred=regressor.predict(X_test)
```

```python
# To called r2_score in sklearn.matrics
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(Y_test,y_pred)
```

```python
# run the r_score
```

```python
r_score
```

```python
# Create the file name to using pickle in .sav
```

```python
import pickle
filename="finalized_model_linear.sav"
```

```python
# Save the model in file
```

```python
pickle.dump(regressor,open(filename,'wb'))
```
